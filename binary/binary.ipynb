```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML
```

```python
provinces_and_territories = ['Alberta',
'British Columbia',
'Manitoba',
'New Brunswick',
'Newfoundland and Labrador',
'Northwest Territories',
'Nova Scotia',
'Nunavut',
'Ontario',
'Prince Edward Island',
'Quebec',
'Saskatchewan',
'Yukon']

def binary_search(arr, item, verbose = False):
    """
    Implements binary search for item in array arr
    Arguments: arr (array to be searched), item (item of interest)
    Returns: Index of item in arr
    """
    
    #Low is zero (index of first item in array)
    #High is len(arr) - 1 (ind)
    #Initial values of variables - they will change!
    low, high = 0, len(arr) - 1
    
    steps = 0
    while low <= high:
        steps += 1
        mid = (low + high) // 2
        guess = arr[mid]
        
        if verbose:
            print(mid)
            print(guess)
        
        if guess == item:
            return mid, steps
        if guess > item:
            high = mid - 1
        else:
            low = mid + 1
            
    return None
        
```

```python
binary_search(provinces_and_territories, 'British Columbia', verbose=True)
```

```python
import matplotlib.pyplot as plt
import random
from statistics import mean 

%matplotlib inline

costs = []

for i in range(100000):
    province_searched = random.choice(provinces_and_territories)
    idx, cost = binary_search(provinces_and_territories, province_searched)
    costs.append(cost)

histplot = plt.hist(costs, bins=range(7))

print(mean(costs))
```

```python
from random import shuffle

def barlist(n):
    bars = [1/float(k) for k in range(1,50)]
    shuffle(bars)
    return bars

fig = plt.figure()

n=100 #Number of frames
x=range(1,50)

barcollection = plt.bar(x,barlist(1))

def animate(i):
    y=barlist(i)
    for i, b in enumerate(barcollection):
        b.set_height(y[i])

anim = animation.FuncAnimation(fig,animate,repeat=False,blit=False,frames=n,
                             interval=100)

```
